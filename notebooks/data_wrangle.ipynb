### Data Exploration Part-1

The first thing that we need to do is to check the properties of the images. we know that CNN networks
need to have images of same size (Height and Weight) and also the mode of the images should also be same which is basically have same dimensions

### Data cleaning part-2

The next thing we need to do is to make sure that the data we have is divided into 3 different folders and classes as
sub directories.
This is how the data structure looks like

- train
  - class_1
  - class_2
  - class_3
  - etc.
- test
  - class_1
  - class_2
  - class_3
  - etc.

1. Raw data consists of 2 directories one that comes with training images and the other with testing. we would be dividing those directories based on the above structure.
2. The other thing that will be processed in here is to convert all the images into same mode which will be RGB, as we noticed from above there are some images that are not of RGB type. 



```python
def create_dirs(abs_loc, dataset_type, labels):
    """
    This will create directory structure needed for the data.
    :param abs_loc: absolute path where the directories needs to be created.
    :type abs_loc: str
    :param dataset_type: dataset type (train or test or val)
    :type dataset_type: str
    :param labels: list of sub directory directory names
    :type labels: list
    :return: None
    """
    # replacing whitespace with _
    label_list = list(map(lambda x: x.replace(' ', '_'), labels))

    for label_name in label_list:
        try:
            os.makedirs(os.path.join(abs_loc, dataset_type, label_name))
        except FileExistsError:
            pass
```

```python
# creating directories for training data set
# replacing whitespaces with _ for dir creation
class_dir_list = list(map(lambda x: x.replace(' ', '_'), data_obj.labels))

# creating directory structure for training dataset.
print('Creating directory structure for training dataset...')
create_dirs(clean_data_path, 'train', class_dir_list)

# creating directory structure for testing dataset
print('Creating directory structure for testing dataset...')
create_dirs(clean_data_path, 'test', class_dir_list)
```

```python
def distribute_images(dataframe, dataset_type, src_dir, dest_dir):
    """
    This will restructure images in their respective sub directories which are 
    based on class labels.
    Note: 
    This method also converts images that are greyscale to rgb type.
    
    :param dataframe: image properties data frame
    :type dataframe: pandas.DataFrame
    :param dataset_type: would be of any of this type, train or test
    :type dataset_type: str
    :param src_dir: source directory where the image is located
    :type src_dir: str
    :param dest_dir: destination directory where the image needs to be stored
    :type dest_dir: str
    :return: None
    """
    
    df_size = len(dataframe)
    success_count = 0
    fail_count = 0
    gray_img_success_count = 0
    gray_img_fail_count = 0
    
    for ind in range(df_size):
        file_to_cpy = os.path.join(src_dir, dataframe['name'][ind])
        dest_loc = os.path.join(dest_dir, dataset_type, dataframe['class'][ind].replace(' ', '_'), dataframe['name'][ind])
        if dataframe['mode'][ind] == 'L':
            try:
                img_obj = Image.open(file_to_cpy).convert('RGB')
                img_obj.save(dest_loc)
                gray_img_success_count += 1
            except:
                gray_img_fail_count += 1
        else:
            if shutil.copy2(file_to_cpy, dest_loc):
                success_count += 1
            else:
                fail_count += 1
        
        print_str = 'processing images... {} of {} processed. {} success {} failed {} gray scale image success {} gray scale image failed'.format(str(ind+1),
                                                                                                                                                  str(df_size),
                                                                                                                                                  str(success_count),
                                                                                                                                                  str(fail_count),
                                                                                                                                                  str(gray_img_success_count),
                                                                                                                                                  str(gray_img_fail_count)
                                                                                       )
        sys.stdout.write('\r' + print_str)
        sys.stdout.flush()
    print('\n')
```

```python
train_raw_data_path = data_obj.train_raw_path
# this will sort the images based on their calss that they belong to
distribute_images(prop_df_train, 'train', train_raw_data_path, clean_data_path)
```

```python
test_raw_data_path = data_obj.test_raw_path
# this will sort the images based on their calss that they belong to
distribute_images(prop_df_test, 'test', test_raw_data_path, clean_data_path)
```
